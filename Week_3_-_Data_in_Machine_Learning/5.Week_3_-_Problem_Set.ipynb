__2. In a markdown box, explain 'underfitting' using the below graph as a reference.__

```python
fig, ax = plt.subplots()
x = np.linspace(10, -10, 100)
y = -x**2 + 4*x + 2  
x = x + np.random.uniform(-3, 3, size=x.shape)
y = y + np.random.uniform(-5, 5, size=y.shape)
ax.scatter(x, y, c='b', label='Training & Testing data')
x_lin = np.array([-10,10])
tmp_f_mb = compute_model_output(x_lin, 0, 0,)
ax.plot(x_lin, tmp_f_mb, c='r', label='ML Model')
ax.set_xticks([])
ax.set_xticklabels([])
ax.set_yticks([])
ax.set_yticklabels([])
plt.legend()
plt.show()
```

Underfitting data occurs when a machine learning model is too simple, failing to learn the relevant patterns within the data. This can be seen in the model above, which has a linear line of best fit despite the data being parabolic in shape

```python
# Visual for questions 4
m = generate_data(10, 55, 10, 55, 200, 0.7)
n = generate_data(0, 50, 0, 40, 50, 0.5)

plt.ylabel("Target")
plt.scatter(m[0], m[1], color='yellow', label='Training Data')
plt.scatter(n[0], n[1], color='black', label='Testing Data')
plt.xticks([])
plt.yticks([])
plt.legend()
plt.show()
```

Bias in machine learning is defined as the difference between the average prediction and the true value. Good machine learning models have consistent, low bias (aka not underfitting), which can be seen in the model above. 

Variance measures the variability of predictions made by a model when trained on different subsets of the same data, reflecting how much the model's predictions change if the data is shuffled. Good machine learning models have low variance (not overfitting).
