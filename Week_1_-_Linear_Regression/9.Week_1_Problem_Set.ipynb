__7. Refering to the below model, describe underfitting in Linear Regression.__

````python
x_train = np.array([1, 1.5, 1.8, 2, 2.2])
y_train = np.array([[200], [250], [260], [280], [500]])
tmp_f_wb = compute_model_output(x_train, 200, 10,)
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')
plt.show()
````

In linear regression, underfitting occurs when a model is too simple, and fails to learn the relevant patterns in the data. As shown in the example above, the algorithm has likely based the line of best fit off the first and last points, creating a steeper gradient which misses a majority of the other points by a large margin.
